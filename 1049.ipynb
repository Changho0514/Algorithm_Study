```python
N, M = map(int, input().split())


set_price = []
piece_price = []
total = 0

for _ in range(M):
    a, b = map(int, input().split())
    set_price.append(a)
    piece_price.append(b)

set_min = min(set_price)
piece_min = min(piece_price)

if set_min < piece_min*6:
    if set_min < (N % 6) * piece_min:
        print((N//6) * set_min + set_min)
    else:
        print((N//6) * set_min + (N%6)*piece_min)

elif set_min >= piece_min*6:
    print(N * piece_min)
```
