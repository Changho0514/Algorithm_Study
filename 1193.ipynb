```python
X=int(input())

num = 0
sum = 0
while sum < X:
    num += 1
    sum += num
sum -= num

if num % 2 == 0:
    i = X - sum
    j = num + 1 - i
else:
    i = num + 1 - (X-sum)
    j = X - sum 

print(f"{i}/{j}")
```
