```python
n = int(input())

count = 0  #  

for i in range(n):
    used = []  #    
    word = input()
    current = word[0]
    used.append(current)
    okay = True
    for j in range(1, len(word)):
        
        if word[j] != current:
            if word[j] in used:
                okay = False
            else:
                current = word[j]
                used.append(current)
    if okay == True:
        count += 1
print(count)
```
