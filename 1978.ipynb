```python
n = int(input())
numbers = map(int, input().split())

count = 0
for num in numbers:
    error = 0
    if num > 1:
        for j in range(2, num): # 2 n-1
            if num % j == 0: 
                error += 1 # 2 n-1   0 error 
        if error ==0:
            count += 1
print(count)
```
