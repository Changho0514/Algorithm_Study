```python
from collections import Counter
import sys

N = int(sys.stdin.readline())
nums = []

for _ in range(N):
    nums.append(int(sys.stdin.readline()))

# 
print(round(sum(nums)/len(nums)))


# 
nums.sort()
print(nums[len(nums)//2])

# 
cnt = Counter(nums).most_common(2)
if len(cnt) > 1:
    if cnt[0][1] == cnt[1][1]:
        print(cnt[1][0])
    else:
        print(cnt[0][0])
else:
    print(cnt[0][0])

# 
print(max(nums) - min(nums))
```
