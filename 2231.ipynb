```python
N = int(input())

if N > 60:
    for i in range(N-60, N+1):
        num = sum((map(int, str(i))))  # i   
        num_sum = i + num
        if num_sum == N:
            print(i)
            break
        if i == N:
            print(0)
if N <= 60:
    for i in range(1, N+1):
        num = sum((map(int, str(i))))  # i   
        num_sum = i + num
        if num_sum == N:
            print(i)
            break
        if i == N:
            print(0)
```

```python
N = int(input())
start_time = time.time_ns()

if N > 60:
    for i in range(N-60, N+1):
        num = sum((map(int, str(i))))  # i   
        num_sum = i + num
        if num_sum == N:
            print(i)
            break
        if i == N:
            print(0)
if N <= 60:
    for i in range(1, N+1):
        num = sum((map(int, str(i))))  # i   
        num_sum = i + num
        if num_sum == N:
            print(i)
            break
        if i == N:
            print(0)

end_time = time.time_ns()

print('  : %20ds' % (end_time - start_time))
```
